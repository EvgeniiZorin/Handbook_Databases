Apache Airflow is a tool for programmatically creating, scheduling, and monitoring workflows. It lets you define pipelines as code (Python), where each pipeline is a DAG:
- Nodes: tasks
- Edges: dependencies (what runs before what)

Functions:
- Workflow orchestrator
- Data pipeline orchestration tool: ETL in DE
- Schedules
- MLOps
- It is open source
