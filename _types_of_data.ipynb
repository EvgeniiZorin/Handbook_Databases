Types of data, by their structure and organisation:

| Data type | Definition | Example |
| - | - | - |
| Structured | Has a well-defined structure, conforms to a data model following a standard order, which is highly organised and easily accessed. Usually it is contained in rows and columns. | SQL databases, CSV |
| Unstructured | Data that generally cannot be contained in a row-column format. There is no pre-defined format or organisation, making it much more difficult to capture, process, and analyse. | Sensor data, emails, blog entries, wikis, word processing documents, PDF files, audio files, videos, images, presentations, web pages. |
| Semi-structured | Are not stored in a relational database, but it has certain organisational properties. | HTML, XML, JSON documents, NoSQL databases |

Metadata - data about data. 
- Author
- file size
- date generated by the document
- Keywords to define the documents
- etc.
